```python
def remove_non_english(text):
    text = re.sub("[^a-zA-Z0-9 ]+", "",text)
    return text
```

```python
def remove_common_strings(x):
    strings = ['restaurant', 'food', 'breakfast', 'lunch', 'dinner', 'snack', 'homemade']
    if type(x) is str:
        for s in strings:
            if s in x:
                x = re.sub(s, '', x)
    return x
```
